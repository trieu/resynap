```python
import pynarrative as pn
import pandas as pd
import altair as alt

# Tạo bộ dữ liệu mẫu: Doanh số bán lẻ hàng năm (triệu đô la)
data = pd.DataFrame({
    'Năm': [2010, 2012, 2014, 2016, 2018, 2020, 2022],  # Năm quan sát
    'Doanh số bán lẻ (Triệu $)': [150, 300, 450, 400, 600, 750, 300]  # Dữ liệu doanh số bán hàng
})

# Tạo biểu đồ cơ bản bằng PyNarrative
chart = pn.Story(data).mark_line().encode(
    x=alt.X('Năm:O', axis=alt.Axis(labelAngle=0)),  # Trục X đại diện cho năm (thang đo thứ tự)
    y='Doanh số bán lẻ (Triệu $):Q'
)

chart.add_title(
    "Tăng trưởng ngành bán lẻ", 
    "2010-2022", 
    title_color="#1a1a1a", 
    subtitle_color="#4a4a4a"
).add_context(
    text=["Xu hướng doanh số bán lẻ qua các năm", 
          "phản ánh tình hình kinh tế và hành vi tiêu dùng.",
          "Biểu đồ thể hiện doanh số bán lẻ hàng năm (triệu đô la)."], 
    position='left', 
    dx=-150,
    color="black"
).render()

chart = chart.add_annotation(
    2022, 400, "Sụt giảm doanh số do Covid-19",
    arrow_direction='left', 
    arrow_dx=50, 
    arrow_dy=-7, 
    arrow_color='red', 
    arrow_size=75,
    label_color='darkgreen', 
    label_size=14,
    show_point=True
).add_next_steps(
    mode='button',
    text="Khám phá thêm",
    url="https://example.com",
    position='bottom',
    title=""
).render()

chart 
```
